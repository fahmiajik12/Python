```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

**Matplotlib.pyplot adalah kumpulan fungsi yang membuat beberapa perubahan pada gambar: misalnya, membuat gambar, membuat area plot dalam gambar, menambah label di plot dan lainnya Biasanya untuk mempermudah secara umum matplotlib.pyplot disingkat menjadi plt import matplotlib.pyplot as plt**

```python
gambar = cv2.imread("iris_mata.jpeg", 0)
cimg = cv2.cvtColor(gambar, cv2.COLOR_GRAY2BGR)
canny = cv2.Canny(cimg, 50, 50)

circles = cv2.HoughCircles(canny, cv2.HOUGH_GRADIENT, 1, 1000, param1=10, param2=15, minRadius=20, maxRadius=40)

circles = np.uint16(np.around(circles))
for i in circles[0, :] :
  #draw the outer circle
  cv2.circle(cimg, (i[0], i[2]), i[1], (0, 255, 0), 2)
  #draw the center of the circle
  cv2.circle(cimg, (i[0], i[2]), 2, (10, 10, 255), 3)
```

**cv2.HoughCircles berfungsi untuk menemukan lingkaran dalam gambar skala abu-abu menggunakan transformasi Hough dan menggunakan Canny Sebagai pendeteksi tepi pada citra mata**
